## W3A16
#### [Rounding]
Nhập vào một số thực: in ra kết quả làm tròn lên ,tròn xuống, làm tròn tới số nguyên gần nhất (không được sử dụng hàm
sẵn có của python). Các kết quả cách nhau một dấu cách 

```python
s=float(input())
def lam_tron(a):
    b=a*10
    if abs(b)%10>=5:
        if a>0:
            return int(a)+1
        else:
            return int(a)-1
    else:
        return int(a)
if s==int(s):
    print(s,s,s)
else:
    if s>0:
        print(int(s)+1,int(s),lam_tron(s))
    else:
        print(int(s),int(s)-1,lam_tron(s))
```
